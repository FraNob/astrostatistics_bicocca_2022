# GPR fit

```python
import numpy as np
import pylab as plt

from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import ConstantKernel, RBF 

from astroML.datasets import generate_mu_z
```

```python
z, mu, dmu = np.asarray(generate_mu_z(100, random_state=1234))

z_sk = z[:,None]
```

```python
# See the effect of changing variance and bandwidth for GPR kernel

#-------  Plt params  ----------------
%config InlineBackend.figure_format='retina'
plt.rc('figure', figsize=(10, 4), dpi=120)
plt.rcParams.update({'font.size': 9})
#--------------------------------------------




z_grid = np.linspace(0,2,200)

fig = plt.figure()

ax = fig.add_subplot(121)

var = 20
l_kern_list = [0.1,0.6,3,10]

for l in l_kern_list:
    kernel = ConstantKernel(var, 'fixed') * RBF(l, 'fixed')
    gpr = GaussianProcessRegressor(kernel = kernel, alpha = dmu**2)
    gpr.fit(z_sk, mu)
    mu_gpr, mu_sigma = gpr.predict(z_grid[:,None], return_std=True)
    
    ax.plot(z_grid, mu_gpr, label= f'length = {l}')
    ax.fill_between(z_grid, mu_gpr - mu_sigma, mu_gpr + mu_sigma, alpha = 0.2)

ax.errorbar(z, mu, dmu, fmt='.k', ecolor='gray', lw=1,label='data', capsize = 2, markersize=5)
ax.set_xlabel(r"$z$")
ax.set_ylabel("$\mu$")
ax.set_xlim(0,2)
ax.set_ylim(35,50)
ax.set_title(f'Variable bw, fixed variance = {var}')
ax.legend()

ax = fig.add_subplot(122)

bw = 3.4
c_kern_list = [0.1,20,1000]

for c in c_kern_list:
    kernel = ConstantKernel(c, 'fixed') * RBF(bw, 'fixed')
    gpr = GaussianProcessRegressor(kernel = kernel, alpha = dmu**2)
    gpr.fit(z_sk, mu)
    mu_gpr, mu_sigma = gpr.predict(z_grid[:,None], return_std=True)
    
    ax.plot(z_grid, mu_gpr, label= f'variance = {c}')
    ax.fill_between(z_grid, mu_gpr - mu_sigma, mu_gpr + mu_sigma, alpha = 0.2)

ax.errorbar(z, mu, dmu, fmt='.k', ecolor='gray', lw=1,label='data', capsize = 2, markersize=5)
ax.set_xlabel(r"$z$")
ax.set_ylabel("$\mu$")
ax.set_xlim(0,2)
ax.set_ylim(35,50)
ax.set_title(f'Variable variance, fixed bw = {bw}')
ax.legend();
```

From the plot on the right, we see that with a reasonable bandwidth, we should worry about having a variance which is high enough. Anyway after a certain point (maybe around 20 here), the change in variance does not have a great influenxe on the best distribution estimate, but it seems crucial to increase uncertainty in poorly - populated areas. **So the parameter search should explore a quite big interval (maybe different orders of magnitude)**

From plot on the left we see that bandwidth is crucial to control the smoothness of the distribution estimate. Anyway we see that with a reasonable choice of variance, passing from 0.1 to 10 we go from overfitting to underfitting quite fast. **The parameter search can stay within thinner bounds.**

```python
# Cross validation using GridSearch, takes a while

# c_kern_list = np.logspace(-2,2,25)
# l_kern_list = np.logspace(-2,2,25)

c_kern_list = np.logspace(-1,4,25)
l_kern_list = np.linspace(0.1,10,25)

k_range = []

for c in c_kern_list:
    for l in l_kern_list:
        
        kernel = ConstantKernel(c, constant_value_bounds='fixed') * RBF(l, length_scale_bounds='fixed')
        k_range.append(kernel)
        
alpha = np.mean(dmu**2)  #  use this as homoscedastic error because gridsearch goes 
                                                    #  crazy if I put the whole array

grid = GridSearchCV(GaussianProcessRegressor(alpha=0.9), {'kernel': k_range}, cv = 10)
grid.fit(z_sk, mu);
```

# Analytic model fit
We will use Nested Sempling to do parameter estimation of the model.

We start with thw model involving dark matter, with $\Omega_m$ and $H_0$ as free parameters

```python
from astropy.cosmology import LambdaCDM
from scipy.stats import norm, uniform

def model(z, h0, om):
    cosmo = LambdaCDM(H0=h0, Om0=om, Ode0=1-om)
    return cosmo.distmod(z).value

def logL(theta):
    ho, om = theta    
    if om<0:
        return -np.inf
    mu_model = model(z, ho, om)
    return np.sum(norm(loc=mu_model, scale=dmu).logpdf(mu))
    
def prior(u):
    x = np.array(u)
    ho_min, ho_max = 40, 100
    om_min, om_max = 0.2, 1
    x[0] = uniform(loc=ho_min, scale=(ho_max - ho_min)).ppf(u[0])
    x[1] = uniform(loc=om_min, scale=(om_max - om_min)).ppf(u[1])
    return x
```

```python
import dynesty

ndim = 2
sampler = dynesty.NestedSampler(logL, prior, ndim, nlive=1000) 
sampler.run_nested()
results = sampler.results
```

```python
results.summary()
```

```python
from dynesty import plotting as dyplot
from dynesty import utils as dyfunc

rfig, raxes = dyplot.runplot(results)
```

```python
labels=['$H_0$', '$\Omega_m$']
tfig, taxes = dyplot.traceplot(results, labels=labels)
```

```python
import corner 

samples = results.samples  
weights = np.exp(results.logwt - results.logz[-1])  

samples_equal = dyfunc.resample_equal(samples, weights)

fig = corner.corner(samples_equal, labels=labels, levels=[0.68,0.95], show_titles=True)
```

```python
# Print medians and 95% quantiles

median_params = []

for i, label in enumerate(labels):
    low, med, sup = np.percentile(samples_equal[:,i],[5,50,95]) 
    print(f'{label}\t=\t{med:.2f}\t+{sup-med:.2f}\t-{med-low:.2f}')
    median_params.append(med)
```

```python
# random samples
N_rand = 30
mask = np.random.choice(len(samples_equal), size = N_rand)
choices = samples_equal[mask]
z_grid = np.linspace(0,2, 200)

for choice in choices:
    plt.plot(z_grid, model(z_grid,*choice), c='C0', alpha = 0.1)
    
plt.plot(z_grid, model(z_grid, 71, 0.27), c = 'k', ls = '--', lw=1, label = 'True values')
plt.errorbar(z, mu, dmu, fmt='.k', ecolor='gray', lw=1,label='data', capsize = 2, markersize=5)
plt.xlabel(r"$z$")
plt.ylabel("$\mu$")
plt.xlim(0,2)
plt.ylim(35,50)
plt.title('Random draws from Nested Sampling evaluation')
plt.legend();
```

We can take a step further and compute not only the best estimate for y from our model, but a pdf($\mu | z$) at every value of $z$.

We have: $$ f(\mu|z) = \int d\theta \ p(\mu |  \theta ; z) \cdot p(\theta) $$

Basically we are averaging the model over the parameters posteriors. 

In practice we can set a grid of z values, sample the distribution by computing model predictions sampling from parameters posteriors, and compute the variance to get dispersion bands similar to GPR

```python
z_gr = np.linspace(0.05, 2, 100)
n_draws = 500

sigma_list = []

for point in z_gr:
    draw_h = np.random.choice(samples_equal.T[0], n_draws)
    draw_o = np.random.choice(samples_equal.T[1], n_draws)

    mu_draw = []

    for h,o in zip(draw_h,draw_o):
        mu_draw.append(model(point,h,o))
    
    sigma_list.append(np.std(mu_draw))
```

```python
sigma_list = np.array(sigma_list)
z_gr = np.linspace(0.05, 2, 100)
    
plt.errorbar(z, mu, dmu, fmt='.k', ecolor='gray', lw=1,label='data', capsize = 2, markersize=5)

mu_analytic = model(z_gr, *median_params)
plt.plot(z_gr, mu_analytic ,  label = 'Best estimate')

plt.fill_between(z_gr, mu_analytic - sigma_list, mu_analytic + sigma_list, color='g', alpha = 0.2, label=r'$1-\sigma$ band')
plt.fill_between(z_gr, mu_analytic - 2*sigma_list, mu_analytic + 2*sigma_list, color='b', alpha = 0.2, label=r'$2-\sigma$ band')


plt.xlabel(r"$z$")
plt.ylabel("$\mu$")
plt.xlim(0,2)
plt.ylim(35,50)
plt.title('Anylitic model fit from Nested Sampling evaluation')
plt.legend();
```

# Mock data generation

First we just visualize how the models would predict a uniform distribution of points in z, then we will generate a mock data sample of 1000 points

```python
# Plotting Predictions from models using a uniform distribution in z

mu_nested=[]
mu_gpr=[]

z_mock_grid = np.random.uniform(0,2,1000)

for zmg in z_mock_grid:
    # Nested sampling
    # To compute the sigma for each point you do int(P(H_0,om | z) * P(mu | params) over params), weight the model with posterior of params
    # Now we are not drawing a gaussian for each x, we are randomly choosing parameters from the posteriors we got from nested sampling
    mask = np.random.choice(len(samples_equal))
    choice = samples_equal[mask]           
    mu_nested.append(model(zmg,*choice)) 
    
    # GPR
    mu_fit, sigma = gpr.predict([[zmg]], return_std=True)
    mu_gpr.append(np.random.normal(loc=mu_fit,scale=sigma))
```

```python
#-------  Plt params  ----------------
%config InlineBackend.figure_format='retina'
plt.rc('figure', figsize=(8, 4), dpi=120)
plt.rcParams.update({'font.size': 9})
#--------------------------------------------

plt.scatter(z_mock_grid,mu_nested,alpha=0.2,label="nested sampling")
plt.scatter(z_mock_grid,mu_gpr,alpha=0.2,label='GPR')
# plt.errorbar(z, mu, dmu, fmt='.k', ecolor='gray', lw=1,label='true data', capsize = 2, markersize=5)


plt.xlabel(r"$z$")
plt.ylabel(r"$\mu$")
plt.xlim(0,2)
plt.ylim(35,50)
plt.title("Predicted points with uncertainty from Nested sampling and GPR")
plt.legend();
```


First we see that $\mu$ and $d\mu$ are correlated:


```python
plt.scatter(z,dmu/mu)
plt.title(r'Relative error vs z')
plt.xlabel(r"$z$")
plt.ylabel(r"$d\mu/\mu$")
mean_rel_err = np.mean(dmu/mu)
print(f'Mean relative error = {mean_rel_err:.3}')
```

We see that the ratio $d\mu/\mu$ is basically constant (with some noise) along z.

To generate mock data we will assign $d\mu$ equal to the product between the mean relative error and the drawn $\mu$.

Now we generate points following statistical properties of data. We will use KDE to get a an estimate pdf for z, then use rejection sampling to get 1000 draws. (Using EmpiricalDistribution(),  artifacts from the coarseness of the histogram appear in mock data)

```python
# Z distribution using KDE, bandwidth set with treshold for KS test p-value of 0.95

from sklearn.neighbors import KernelDensity

kde = KernelDensity(bandwidth = 0.07,  kernel='gaussian')

kde.fit(z[:, None])   

def kde_pdf(x):
    arr = np.asarray(x)
    logl = kde.score_samples(arr[:,None])
    return np.exp(logl)

plt.plot(z_grid,kde_pdf(z_grid))
plt.title(r'pdf of z using KDE')
plt.xlabel(r"$z$")
plt.ylabel(r"pdf");
```

```python
from scipy.stats import ks_2samp
rng = np.random.default_rng(404) 

N = 10000 # number of draws
ymax = 1.7
xmax = 2.0
x = rng.uniform(0, xmax, N)
y = rng.uniform(0, ymax, N)

good_points = x[y<=kde_pdf(x)]

# z_upsampled = np.concatenate((z,good_points))
z_mock = good_points[:1000]

D, p = ks_2samp(z, z_mock)
print("KS test: D = %.2g; p = %.2g" % (D, p))

# Plot
fig, ax = plt.subplots()

hist_true = ax.hist(z,bins='auto',density=True, histtype='step', label='original')
hist = ax.hist(z_mock,bins='auto',density=True, histtype='step', label='cloned')
ax.legend();
```
